<h1>Process of the dataset creation.</h1>

<h3>1.Sign in to account Copernicus Browser</h3>

<h3>2.Select the necessary settings</h3>

In my case i used:

Max. cloud coverage:20%


Data collection: Sentinel-2 L2A

<h3>3.Create an area of interest</h3>

Draw polygonal area or rectangular area

<h3>4.Choose dataset</h3>


Press find products for current view and choose suitable. I used data for 06/06/25 and 27/09/25.

For my task I chose an area that includes fields and a river. I used images taken in different seasons so that the algorithm would be resistant to changes in vegetation, light, and clouds.

<h3>5.Download dataset</h3>

I get compressed folder like "S2B_MSIL2A_20250606T082559_N0511_R021_T37UCQ_20250606T112504.SAFE". For my task I need to unpack it and extract the necessary parameters.
